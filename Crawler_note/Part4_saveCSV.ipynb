我們預計可以到https://www.taifex.com.tw/eng/eng3/eng3_2dl.asp , 並且把他想要的資訊送進去。

So,先把data弄出來

```python
test = "goday=&DATA_DATE=2015%2F04%2F01&DATA_DATE1=2015%2F04%2F09&DATA_DATE_Y=2015&DATA_DATE_M=04&DATA_DATE_D=01&DATA_DATE_Y1=2015&DATA_DATE_M1=04&DATA_DATE_D1=09&syear=2015&smonth=04&sday=01&syear1=2015&smonth1=04&sday1=09&datestart=2015%2F04%2F01&dateend=2015%2F04%2F09&COMMODITY_ID=all&commodity_id2t=&his_year=2014"
```

```python
test = test.replace("%2F","/")
```

```python
data = dict(map(lambda xx:tuple(xx.split("=")),test.split("&")))
```

```python
data
```

##data是我們要的資料
requests.post是可以把資訊送進去再回傳

```python
res=requests.post("https://www.taifex.com.tw/eng/eng3/eng3_2dl.asp",data=data)
```

用history可以知道之前發生的事

```python
res.history #是一個list
```
