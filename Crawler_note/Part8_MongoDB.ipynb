#存CSS Selector 到 MongoDB

```python
import requests
```

```python
res=requests.get("http://www.w3schools.com/cssref/css_selectors.asp")
```

```python
res
```

```python
res.encoding
```

```python
res.encoding='utf-8'
```

```python
import pandas as pd
```

```python
res.text
```

```python
tables=pd.read_html(res.text)
```

```python
len(tables)
```

```python
tables[0].iloc[:,:3]
```

```python
tb=tables[0].iloc[:,:3]
```

```python
tb.columns=tb.ix[0]
```

```python
tb=tb[1:]
```

```python
tb
```

```python
data=tb.to_dict(orient="records")
```

```python
tb.to_csv('css_selector.csv',encoding='utf-8')
```

##重頭戲！！！存進MongoDB
MongoDB是一種NoSQL的DataBase

引入pymongo套件

```python
from pymongo import MongoClient
```

接上主db

```python
connection=MongoClient("mongodb://localhost:27017")
```

連上要的db

```python
db=connection.test
```

再接上要的table

```python
cssdata=db.cssdata
```

把instance insert進來

```python
cssdata.insert(data)
```
